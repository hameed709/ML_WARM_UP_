```python
import pandas as pd
```

```python
data={
    "Name":["Abdul","Ayaan","Sara","Fatima"],
    "Score":[85,90,78,92]
}
df=pd.DataFrame(data)
df.to_csv("student_scores.csv",index=False)
```

```python
df=pd.read_csv("student_scores.csv")
print(df)
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
avg_score=df['Score'].mean()
print(f'Average score : {avg_score}')
```
