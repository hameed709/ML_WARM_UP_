```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

```python

x_pred=[0,0,0,1,1,1]
y_pred=[0,1,1,0,1,0]
```

```python
print(f"Confusion Matrix : \n{confusion_matrix(x_pred,y_pred)}")
print(f"Precision : {precision_score(x_pred,y_pred)}")
print(f"Recall : {recall_score(x_pred,y_pred)}")
print(f"F1-Score : {f1_score(x_pred,y_pred)}")
```
