```python
import pandas as pd

df = pd.read_csv("student_scores.csv")
print(df)
```

```python
df.loc[2, 'Score'] = None
print(df)
```

```python
print(df.isnull())        # True/False matrix
print(df.isnull().sum())  # Count missing per column
```

```python
df['Score'].fillna(df['Score'].mean(), inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.loc[len(df)] = ['Abdul', 85]
```

```python
print(df.duplicated())
print(df[df.duplicated()])
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.rename(columns={'Name': 'Student_Name', 'Score': 'Final_Score'}, inplace=True)
print(df.head())
```

```python
print(df.describe())  # statistics summary
print(df.info())      # data types and null count
print(df.shape)       # rows x columns
```

```python
df.to_csv("student_scores_cleaned.csv", index=False)
print("✅ Cleaned dataset saved successfully!")
```
