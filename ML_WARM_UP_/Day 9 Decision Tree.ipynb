```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
X=[[1],[2],[3],[4],[5],[6]]
y=[0,0,0,1,1,1]
```

```python
model=DecisionTreeClassifier(max_depth=2)
model.fit(X,y)
```

```python
plt.figure(figsize=(8,6))
plot_tree(model,filled=True)
plt.show()
```
