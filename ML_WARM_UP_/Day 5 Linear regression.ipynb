```python
# linear_manual.py
hours = [1, 2, 3, 4, 5]
marks = [40, 50, 60, 70, 80]

# assume line y = m*x + b
m = 11
b = 31

predictions = [m*h + b for h in hours]

# mean squared error
mse = sum((p - y)**2 for p, y in zip(predictions, marks)) / len(marks)

print("Hours:", hours)
print("Actual marks:", marks)
print("Predictions:", predictions)
print(f"MSE: {mse:.3f}")
```
